```python
import zipfile
import os

def folder_to_zip(folder_path, zip_path):
    # zip 파일 생성 (쓰기 모드)
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # 폴더 내 모든 파일과 하위 폴더를 순회
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                # 파일의 전체 경로
                file_path = os.path.join(root, file)
                # zip 내에서 사용할 상대 경로
                arcname = os.path.relpath(file_path, folder_path)
                # zip 파일에 추가
                zipf.write(file_path, arcname)

# 사용 예시
folder_to_zip('/content/data', '/content/drive/MyDrive/컴퓨터비전/포폴 프로젝트/주차장시스템/데이터/data ocr/ocr_dataset.zip')
```
