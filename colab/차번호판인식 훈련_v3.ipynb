# 데이터 분리

```python
from glob import glob

img_list = glob('/content/data/images/*.jpg')
txt_list = glob('/content/data/labels/*.txt')

print(len(img_list), len(txt_list))
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# 데이터와 라벨이 있는 폴더 경로
data_dir= "/content/data/images"
label_dir = "/content/data/labels"

# 출력 경로 설정
train_images_dir = "/content/export/train/images"
train_labels_dir = "/content/export/train/labels"
val_images_dir = "/content/export/val/images"
val_labels_dir = "/content/export/val/labels"

# train/test 비율
test_size = 0.2

# 폴더 생성
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# 데이터와 라벨 파일 리스트 가져오기
data_files = sorted([f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir, f))])
label_files = sorted([f for f in os.listdir(label_dir) if os.path.isfile(os.path.join(label_dir, f))])

# 데이터와 라벨 이름 일치 확인
if len(data_files) != len(label_files):
  raise ValueError("데이터 파일과 라벨 파일의 개수가 다릅니다.")
if any(data_file.split('.')[0] != label_file.split('.')[0] for data_file, label_file in zip(data_files, label_files)):
    raise ValueError("데이터 파일과 라벨 파일의 이름이 일치하지 않습니다.")

# 데이터 분
train_data, val_data, train_labels, val_labels = train_test_split(
    data_files, label_files, test_size=test_size, random_state=42
)

# 파일 복사
def copy_files(file_list, src_dir, dest_dir):
    for file in file_list:
        shutil.copy(os.path.join(src_dir, file), os.path.join(dest_dir, file))

# 훈련 데이터 복사
copy_files(train_data, data_dir, train_images_dir)
copy_files(train_labels, label_dir, train_labels_dir)

# 테스트 데이터 복사
copy_files(val_data, data_dir, val_images_dir)
copy_files(val_labels, label_dir, val_labels_dir)
```

```python
import zipfile
import os

def folder_to_zip(folder_path, zip_path):
    # zip 파일 생성 (쓰기 모드)
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # 폴더 내 모든 파일과 하위 폴더를 순회
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                # 파일의 전체 경로
                file_path = os.path.join(root, file)
                # zip 내에서 사용할 상대 경로
                arcname = os.path.relpath(file_path, folder_path)
                # zip 파일에 추가
                zipf.write(file_path, arcname)

# 사용 예시
folder_to_zip('/content/export', '/content/drive/MyDrive/컴퓨터비전/포폴 프로젝트/주차장시스템/데이터/data v2/dataset.zip')
```
